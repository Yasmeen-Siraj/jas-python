```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup as bs
import pandas as pd

import sqlite3
```

```python
# Load the Web page

chrome_path = r"C:/Program Files (x86)/chromedriver.exe"
driver = webdriver.Chrome(chrome_path)
driver.get("http://www.jodidb.org/TableViewer/tableView.aspx?ReportId=93906")

```

```python
#Specify the Drop Down Menu

drop_menu = driver.find_element_by_xpath('//*[@id="selectMenu2"]').click()
```

```python
# Specify the Exports option from the menu 

ex_obj=driver.find_element_by_xpath('//*[@id="a-el-d2-mi3"]').send_keys("Exports")
```

```python
# Click Exports

ex_obj2= driver.find_element_by_xpath('//*[@id="a-el-d2-mi3"]').click()
```

```python
# Extract teh columns headers
```

```python
src = driver.page_source
parser = bs(src, "lxml")
table = parser.find("table", attrs={"class": "TVDataTable"})

#driver.find_elements_by_class_name('TVItemColHeader')
#headers = table.findAll('th')
headers = driver.find_elements_by_class_name('TVItemColHeader' )
headerlist = [h.text.strip()  for h in headers[0:]]
```

```python
headerlist
```

```python
#Extract the rows headers
```

```python

row_headers = driver.find_elements_by_class_name('TVItemRowHeader')
row_list = [r.text.strip() for r in row_headers[0:]]
```

```python
row_list
```

```python
rows= table.findAll('tr')[2:]
```

```python
cell_data = [[td.getText().strip() for td in rows[i].findAll('td')[0:]] for i in range(len(rows))]
```

```python
len(rows)
```

```python
cell_data
```

```python
# Create the dataframe
```

```python
stat = pd.DataFrame(cell_data, columns=headerlist)
```

```python
stat
```

```python
stat
```

```python
stat.insert(0, 'Country', row_list)
stat
```

```python
stat.count()
```

```python
# Convert dataframe to SQL data
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite://', echo = False)
```

```python
stat.to_sql('Country_Balance', con=engine)
```

```python
print(engine.execute("SELECT * FROM Country_Balance").fetchall())
```

```python
stat2 = pd.read_sql('Country_Balance', con= engine)
```
