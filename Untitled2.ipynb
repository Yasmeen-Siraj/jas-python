```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("http://www.jodidb.org/TableViewer/tableView.aspx?ReportId=93906")
```

```python
soup = bs(r.content)
```

```python
print(soup.prettify())
```

```python
table1= soup.find("table", attrs={"id": "MainTable"})
data_table = table1.find("table", attrs={"id": "DataTable"})
print(data_table.prettify())
```

```python
table_head = soup.find("thead")
print(table_head.prettify())
```

```python
col_header = table_head.find("th", attrs={"id" : "id=a0"})
print(col_header)
```
